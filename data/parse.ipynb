```python
import numpy as np
import pandas as pd
import json
```

```python
df_liv = pd.read_csv('./live.csv', date_parser='timestamp')
df_ref = pd.read_csv('./reference.csv')
```

```python
df_liv
```

```python
cat_list = df_ref.label.unique()
cat_list.sort()
cat_list
```

```python
df_ref['cat_id'] = df_ref.apply(
    lambda x: np.where(cat_list == x.label)[0][0], axis=1
)
```

```python
df_liv['cat_id'] = df_liv.apply(
    lambda x: np.where(cat_list == x.prediction)[0][0], axis=1
)
```

```python
df_liv
```

```python
list_liv = []
for i in df_liv.index:
    dp = df_liv.loc[i]
    list_liv.append([
       [float(dp[0]), float(dp[1])],
        int(dp.cat_id),
        str(dp.timestamp)
    ])

```

```python
OUT_FOLDER = './'
with open("{}list_liv.json".format(OUT_FOLDER), "w") as json_file:
    #json_list = json.dumps(list_liv)
    #print(json_list)
    json.dump(list_liv, json_file)
```
